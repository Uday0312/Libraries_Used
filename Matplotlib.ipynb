```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt

%matplotlib inline
```

SIN AND COS GRAPH

```python
x = np.linspace(0, 4*np.pi, 100)

sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x,sin_y,label='Sine')
plt.plot(x,cos_y,label='Cos')

plt.legend(loc='upper right')
plt.title('Sine Wave')

plt.show()

```

Different Kind of Plots

```python
Line Plot = two continuous variables.
```

```python
x = np.linspace(0,4*np.pi, 100)
y = np.sin(x)

plt.style.use('Solarize_Light2')

plt.plot(x, y)

plt.title('Sine & Cosine Wave')

plt.show()
```

```python
x= np.linspace(0,4*np.pi,100)

y_sin=np.sin(x)
y_cos=np.cos(x)

plt.figure(figsize=(15,5))
plt.style.use('Solarize_Light2')

plt.plot(x, y_sin,  '*' ,color = 'red' , linewidth = 1, label = 'sin')
plt.plot(x, y_cos, '--' , color = 'blue', linewidth = 1, label = 'cos')


plt.plot('S&C Wave')
plt.legend(loc= 'upper right')
plt.title('S&C WAVE')
plt.show()
```

```python
x= np.linspace(0,20,21,dtype = 'int')
plt.figure(figsize=(15,8))

plt.plot(x, x + 0, linestyle = 'solid'  , label = 'Solid')
plt.plot(x, x + 1, linestyle = 'dashed' , label = 'Dashed')
plt.plot(x, x + 2, linestyle = 'dashdot', label = 'DashDot')
plt.plot(x, x + 3, linestyle = 'dotted' , label = 'Dotted')

plt.legend(loc = 'lower right')

plt.show()
```

**SCATTER PLOTS**- individual data points are represented as points in a two-dimensional plane

```python
n=500
x=np.random.randn(n)
y=np.random.randn(n)
plt.figure(figsize=(15,8))
plt.scatter(x,y,c='green',alpha=0.7,s= 100)

plt.title('Scatter Plot')
plt.legend( loc = 'upper right')
plt.show()
```

In this code, the scatter function is used to create a scatter plot with 500 data points, represented as red markers with a size of 100  and a transparency of 0.7

Multiple Scatter Plots

```python
n=500

x1=np.random.randn(n)
y1=np.random.randn(n)

x2=np.random.randn(n)+2
y2=np.random.randn(n)+2

plt.scatter(x1,y1, c='green',alpha=0.9,s=10,label='swiggy')
plt.scatter(x2,y2, c='blue',alpha=0.5,s=10,label='Zomato')

plt.title('Scatter with Multiple Groups')
plt.legend(loc= 'upper right')
plt.show()
```

**HISTOGRAMS-set of continuous or discrete data**




```python
data=np.random.randn(1000)

plt.hist(data)

plt.show()
```

```python
data=np.random.randn(10000000)

plt.hist(data, bins=2000,edgecolour='black',alpha=0.9,color='blue')

plt.title('Random Hist')
plt.ylabel('Frequency')
plt.xlabel('values')

plt.show()
```

**MULTIPLE HIS**

```python
x1=np.random.randn(1000)
x2=np.random.randn(1000) + 2
x3=np.random.randn(1000) + 4

plt.hist(x1,bins=20,alpha=0.8)
plt.hist(x2,bins=20,alpha=0.8)
plt.hist(x3,bins=20,alpha=0.8)

plt.title('Random Histogram')
plt.ylabel('Frequency')
plt.xlabel('Values')

plt.show()
```

 **PIECHART**-

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]

fig = plt.figure(figsize = (4,4))

plt.pie(data, labels = cars)
plt.title('Distribution of cars')
plt.show()
```

**Basic Operations**

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data=[450  , 385 , 35   ,   10  ,  200   , 500]
explode=[0,0,0,0,.1,0]

fig= plt.figure(figsize=(4,4))

plt.pie(data,labels=cars,explode = explode)
plt.show()
```

**Shadows**

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data=[450  , 385 , 35   ,   10  ,  200   , 500]
explode=[0,0,0,0,.1,0]

fig= plt.figure(figsize=(4,4))

plt.pie(data,labels=cars,explode = explode,shadow=True)
plt.show()
```

**ADDING PERCENTAGE**

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   50  ,  200   , 500 ]

fig = plt.figure(figsize=(4,4))

plt.pie(data, labels=cars, autopct='%2.2f%%')

plt.show()
```

**Multiple** **Subplots**

```python
x = np.linspace(0,10,50)

fig, ax = plt.subplots(2,2)

y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sin Wave')


y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('Cos Wave')

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('Tan Wave')


y = np.random.randn(1000)
ax[1,1].hist(y,bins = 40)
ax[1,1].set_title('Random Distribution')


plt.tight_layout()
plt.show()
```
